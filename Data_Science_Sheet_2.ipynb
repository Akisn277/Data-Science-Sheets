EMPLOYEE RECORDS- Data cleaning, dealing with NaN, Updating col name

```python
import pandas as pd
```

```python
emp_rec={
    "Emp_id":[101,102,103,104,105],
    "Name":["Aki","Indhu","Niha","Aathu","Steph"],
    "Designation":["Developer","CEO","Engineer","CFO","Manager"],
    "Salary":[1000000,10,10000,200000,30000],
    "Experience":[3,4,5,6,6],
    "Age":[20,21,22,23,24]
}

df=pd.DataFrame(emp_rec)
print(df)
```

```python
df_new=df[df['Salary'] >= 10000]
print(df_new)
```

```python
new_data= {
    "Emp_id":[106],
    "Name":["Cindy"],
    "Salary":[75000],
    "Experience":[2],
    "Age":[28]
}

df_newrow=pd.DataFrame(new_data)
df_updated = pd.concat([df, df_newrow], ignore_index=True)
print(df_updated)
```

```python
df_updated['Designation'] = df_updated['Designation'].fillna('Unknown')
display(df_updated)
```

```python
df_updated=df_updated.rename(columns={'Experience':'Exp'})
display(df_updated)
```

```python
df_updated['Age']=df_updated['Age'] + 3
display(df_updated)
```

STUDENT RECORDS- Merge(inner,outer,left,right),Concat

```python
student_data = {
    'Roll Num': [1, 2, 3, 4, 5],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve'],
    'Subject1_Marks': [85, 78, 92, 65, 88],
    'Subject2_Marks': [70, 82, 89, 75, 95]
}

df_students1 = pd.DataFrame(student_data)
print(df_students1.head())
```

```python
student_data2 = {
    'Roll Num': [3, 4, 5, 6, 7],
    'Name': ['Charlie', 'David', 'Eve', 'Frank', 'Grace'],
    'Subject3_Marks': [70, 85, 90, 75, 80],
    'Subject4_Marks': [60, 70, 80, 85, 90]
}

df_students2 = pd.DataFrame(student_data2)
print(df_students2.head())
```

```python
df_inner_merge = pd.merge(df_students1, df_students2, on='Roll Num', how='inner')
print(df_inner_merge)
```

```python
df_outer_merge = pd.merge(df_students1, df_students2, on='Roll Num', how='outer')
print(df_outer_merge)
```

```python
df_right_merge = pd.merge(df_students1, df_students2, on='Roll Num', how='right')
print(df_right_merge)
```

SALES RECORDS- Pivoting and melting

```python
data = {
    'Region': ['East', 'West', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'B'],
    'Month': ['Jan', 'Jan', 'Jan', 'Jan', 'Feb', 'Feb'],
    'Sales': [100, 150, 200, 120, 110, 180]
}
df_sales = pd.DataFrame(data)
display(df_sales)
```

### Pivot Operation

The `pivot` function is used to reshape data (produce a 'pivot' table) based on column values. It takes three main arguments:
- `index`: Column to use to make new frame's index.
- `columns`: Column to use to make new frame's columns.
- `values`: Column(s) to use for populating new frame's values.

Let's pivot our `df_sales` to see `Sales` for each `Region` across different `Product` types.

```python
df_pivot = df_sales.pivot_table(index=['Month', 'Region'], columns='Product', values='Sales', aggfunc='sum')
display(df_pivot)
```

### Melt Operation

The `melt` function is the inverse of `pivot`. It transforms or 'unpivots' DataFrames from a wide format to a long format. It's useful when you want to convert columns into rows.
- `id_vars`: Columns to use as identifier variables.
- `value_vars`: Columns to unpivot. If not specified, all columns not set as `id_vars` will be unpivoted.
- `var_name`: Name for the new 'variable' column.
- `value_name`: Name for the new 'value' column.

Let's melt our original `df_sales` DataFrame, treating `Region`, `Product`, and `Month` as identifier variables and `Sales` as the value variable. This might seem redundant for `df_sales` as it's already in a long format, but it's a good way to understand the concept. A more common use case for melt would be to unpivot the `df_pivot` DataFrame.

```python
df_melted_original = df_sales.melt(id_vars=['Region', 'Product', 'Month'], value_vars=['Sales'], var_name='Metric', value_name='Amount')
display(df_melted_original)
```

```python
df_melted_from_pivot = df_pivot.reset_index().melt(id_vars=['Month', 'Region'], var_name='Product', value_name='Sales')
display(df_melted_from_pivot)
```
